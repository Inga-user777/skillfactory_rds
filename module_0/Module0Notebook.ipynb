```python
import numpy as np

def get_random_numbers(range_start,range_end,num_qty=10):
    '''' 
    Generates random numbers in a specific range
    If num_qty does not have value then it takes 10 by default        
    ''''
    np.random.seed(1)    
    return np.random.randint(range_start,range_end+1, size=(num_qty))


def find_the_number(range_start, range_end, the_number):
    '''' 
    Looks for a value in a loop by cutting off ranges
    which don't match already calculated conditions
    Can find value in log2(n) tries
    '''' 
    counter = 0    
    while True:
        counter += 1
        current = (range_start+range_end)//2 
        if current == the_number:            
            break
        else:
            if the_number > current:
                range_start = current+1
            else:
                range_end = current-1
    return counter


def check_algoritm_efficiency(range_start, range_end, num_of_iterations):
    '''' 
    Executes num_of_iterations times and returns arithmetic mean of values
    '''' 
    res = []
    
    for number in get_random_numbers(range_start,range_end,num_of_iterations):
        res.append(find_the_number(range_start,range_end,number))        
    
    return np.mean(res)


def main():
    while True:
        range_start = int(input("Please input range start: "))
        if range_start < 0:
            print("Range start cannot be negative number.")
        else:
            break
    
    while True:
        range_end = int(input("Please input range end: "))
        if range_end < 0:
            print("Range end cannot be negative number.")
        elif range_end <= range_start:
            print("Range end must be greater than range start.")
        else:
            break
    
    while True:
        num_of_iterations = int(input("Please input number of iterations\ 
        you would like to run: "))
        if num_of_iterations <= 0:
            print("Number of iterations must be greater than 0.")
        else:
            break

    result=check_algoritm_efficiency(range_start,range_end,num_of_iterations)
    print(f"Alghoritm guesses the number in {result} time(s) on avarage.")


if __name__== "__main__" :
    main()       
```
